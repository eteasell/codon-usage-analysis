```python
from pathlib import Path
import sys
sys.path.append(str(Path().resolve().parent))

import pandas as pd
import matplotlib.pyplot as plt

path = '../data/dependent_hydrophobic_genes.csv'
df = pd.read_csv(path, index_col='gene')
print(df)

start = df['first_hydrophobic_start']
end = df['first_hydrophobic_end']
lengths = end - start
```

```python
plt.hist(end, bins=30, edgecolor='black')
plt.xlabel('Targeting Sequence End Position')
plt.ylabel('Frequency')
plt.title('Targeting Sequence End Positions in SRP-dependent Proteins')
plt.show()
```

```python
plt.hist(lengths, edgecolor='black')
plt.xlabel('Length of Targeting Sequence (amino acids)')
plt.ylabel('Frequency')
plt.title('Targeting Sequence Lengths in SRP-dependent Proteins')
plt.show()
```
